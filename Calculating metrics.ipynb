## 4. Calculate metrics

```python
all_targets=list(pd.unique(All_result['Target']))
```

```python
all_targets
```

```python
all_folds=list(pd.unique(All_result['Fold']))
```

```python
all_folds
```

```python
all_models=list(pd.unique(All_result['Model']))
```

```python
all_models
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
target_dict= {
    'apa': 'After putting away',
    'bon' : 'Behaviour on hand',
    'spos' : 'Sleep position',
    'spoz' : 'sleep poze',
    'sr' :  'Splashing reaction',
    'sr2' : 'Splashing reaction 2',
    't': 'Touch reaction'}
```

```python
target_dict['apa']
```

```python
metrics_to_save=pd.DataFrame()
for x in all_models: #Model
    for y in all_targets: # Target
        fold_scores_f1=[]
        fold_scores_prec=[]
        fold_scores_recall=[]
        fold_scores_accuracy=[]
        for z in all_folds: #Fold
            data_silce = All_result.loc[(All_result['Model'] == x) & (All_result['Target'] == y) & (All_result['Fold'] == z)]
            pred = data_silce['pred']
            obs = data_silce['obs']
            
            F_score_f1 = f1_score(obs, pred, average="macro")
            F_score_precison = precision_score(obs, pred, average="macro")
            F_score_recall = recall_score(obs, pred, average="macro")
            F_score_accuracy = accuracy_score(obs, pred)
            
            fold_scores_f1.append(F_score_f1)
            fold_scores_prec.append(F_score_precison)
            fold_scores_recall.append(F_score_recall)
            fold_scores_accuracy.append(F_score_accuracy)
                                        
        A_F_score_f1=sum(fold_scores_f1)/len(fold_scores_f1)
        A_F_score_precison=sum(fold_scores_prec)/len(fold_scores_prec)
        A_F_score_recall=sum(fold_scores_recall)/len(fold_scores_recall)
        A_F_score_accuracy=sum(fold_scores_accuracy)/len(fold_scores_accuracy)
        
        data_fold=pd.DataFrame({"Model" : [x],
                                   "Target" : target_dict[y],
                                   "F1": [A_F_score_f1],
                                   "Precision": [A_F_score_precison],
                                   "Recall": [A_F_score_recall],
                                   "Accuracy": [A_F_score_accuracy]})
        metrics_to_save=pd.concat([metrics_to_save,data_fold])
```
